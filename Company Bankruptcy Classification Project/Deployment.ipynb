### **5. Deployment of the fitted Model**

```python
#Importing the packages
import streamlit as st
import pandas as pd
import numpy as np
import pickle 
import requests
from io import BytesIO
```

```python
#!pip install streamlit
```

```python
#Loading the model
model_url = 'https://raw.githubusercontent.com/JackronyK/Projects_Jackrony/blob/main/Company%20Bankruptcy%20Classification%20Project/Models/RandomForest.pkl'

#Function to load the model from a url
def load_model(url):
    response = requests.get(url)
    response.raise_for_status #Cheking if the request was successful
    model_file = BytesIO(response.content)
    model = pickle.load(model_file)
    return model

#Loading the model
model = load_model(model_url)
#Streamlit app title
st.title('Bankrupty Prediction Model')

#File uploader for user to upload a csv file
upload_file = st.file_uploader('Upload a csv containing the features')

if upload_file is not None:
    #read the file
    data = pd.read_csv(upload_file)

    #Display the uploaded file
    st.write('Uploaded data: ')
    st.write(data)

    #Predict usnig the Loaded model
    predictions = model.predict(data)

    #adding the prediction to the data file
    data['predictions'] = predictions

    #Display the prediction
    st.write('Predictions: ')
    st.write(data)
else:
    st.write('Please Upload a csv file')

```
